**shuffling the order of Tensors**===(so why might this be neccessary)
--
suppose we want to train a Neural Network to identiffy some things from a huge data set if we pass in the first category and then the second ,the neural network might start adapting to the first categories pattterns only to be shocked when it realises it has to learn category 2 also .conversely if the groups are mixed beforehand then the NeuralNetwork will have to figure them out by itself and therefore not get spooked by happenings .

```python
##Shuffling ;so that the inherrent order doesn't affect learning
shuffle=tf.constant([[3,4],[5,6],[9,4]])

tf.random.shuffle(shuffle)
```

```python
tf.random.set_seed(42)
tf.random.shuffle(shuffle,seed=43)
```

#**What is a seed in TF **
A seed" in TensorFlo "seed" in TensorFlow is a number you set to make random actions in your code predictable and repeatable. It ensures that if you run your code today, tomorrow, or a year from now, and someone else runs it on their computer, as long as they use the same seed, they'll get the same "random" outcomes, and thus, more consistent and reproducible results for your model training and experiments.


>if we want to shuffle our tensors according to some pattern ,we use the global as well as operation seed of the same value
--
this is crucial if we want to have a shuffle in a discrete order

##one hot_encodding

method of representing/converting categorical data in numerical format

.easier for neural networks

.compartibility with other algorithms

.performance

---
tf.one_hot(),

on value & off values allow you to specify what values will be displayed on either of the situations,not neccesarily 1 and 0 as is the norm .

```python
#create a list of some indices
some_list=[1,2,3,4]  #let's say they represent red,green,blue and white
some_list_1=[0,1,2,3]
some_list_2=[2,3,4,5]
```

```python
print("\narray [0,1,2,3]:\n",tf.one_hot(some_list_1,depth=3))
print("\narray [1,2,3,4]:\n",tf.one_hot(some_list,depth=3))
print("\narray [2,3,4,5]:\n",tf.one_hot(some_list_2,depth=3))
print("\n------------------------------------------\n")
print("array [0,1,2,3]\n",tf.one_hot(some_list_1,depth=4))
print("array [0,1,2,3]\n",tf.one_hot(some_list,depth=4))
print("array [0,1,2,3]\n",tf.one_hot(some_list_2,depth=4))
```
