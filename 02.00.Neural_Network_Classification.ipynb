##What is neural network classification ?
A type of learning where an model predicts a class/category of a given data

---
Types of classification

1. Binary classification: Predicts between two possible classes, such as "yes/no," "true/false," or "spam/not spam".
2. Multiclass classification: Predicts among three or more possible classes, such as classifying an image as a "cat," "dog," or "bird".
3. Multi-label classification: Assigns multiple labels to a single instance, unlike multiclass classification where only one label is predicted per instance.

---
Examples of classification problems

1. Spam detection: Classifying emails as either "spam" or "not spam".
2. Image recognition: Identifying whether an image contains a "dog" or a "cat".
3. Medical diagnosis: Predicting if a patient has a certain disease or not.
4. Sentiment analysis: Determining if a piece of text expresses a positive, negative, or neutral sentiment.
5. Fraud detection: Classifying a transaction as either fraudulent or legitimate.

The sklearn.datasets.make_circles function in scikit-learn generates a synthetic dataset specifically designed for binary classification problems where the data points are arranged in two concentric circles.

In simple terms, it creates a "bullseye" pattern:
Two Classes: It generates two distinct groups of data points, representing two different classes (e.g., class 0 and class 1).
Concentric Circles: One class forms the inner circle, and the other class forms the outer circle.

Non-linear Separability: This arrangement makes the dataset linearly inseparable, meaning a straight line cannot effectively separate the two classes. This challenges simple linear models and highlights the need for more complex, non-linear classification algorithms.
Noise Control: You can add a controlled amount of noise to the data, making the separation less perfect and more realistic.

Why is this useful?

make_circles provides a simple, visual, and challenging dataset to:
Test and illustrate classification algorithms: It helps demonstrate how different algorithms, especially those capable of handling non-linear decision boundaries (like Support Vector Machines with non-linear kernels, or neural networks), perform in a non-linear setting.

Understand the limitations of linear models: It clearly shows why linear models would struggle with this type of data.
Visualize model performance: The circular nature of the data makes it easy to visualize the decision boundary learned by a classification mode

Regression loss functions focus on minimizing the numerical distance between predicted and actual continuous values,**while classification loss functions focus on minimizing the discrepancy between predicted class probabilities and actual class labels. This difference in objective dictates the appropriate loss function choice for a given task.**

---
Classification Loss Functions:

1. Purpose: Used when the neural network predicts a discrete category or class. Examples include classifying emails as spam or not spam, identifying objects in an image, or determining if a tumor is benign or malignant.

2. Measurement: Quantify the discrepancy between the predicted probability distribution over classes and the true class label.
Common Examples:
 * **Binary Cross-Entropy (BCE)**: Used for binary classification problems (two classes). It measures the difference between the predicted probability of belonging to the positive class and the true binary label.

* **Categorical Cross-Entropy**: Used for multi-class classification problems (more than two classes). It measures the difference between the predicted probability distribution over all classes and the one-hot encoded true class label.

* **Sparse Categorical Cross-Entropy**: Similar to categorical cross-entropy but used when the true class labels are integers rather than one-hot encoded vectors.

* **Hinge Loss**: Primarily used in Support Vector Machines (SVMs) and can also be applied to neural networks for classification, particularly for binary classification. It aims to maximize the margin between classes.

##We need to visualize the predictions of our model
1. We need a function to take in the x & y values of the model
2. create a meshgrid of different values
3. Make predictions across the meshgrid
4. PLot the predictions & line between grid zone

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_decision_boundary(model,x,y):
  """
  Takes in the model & features and labels and plots the preds

  the meshgrid has the +_(0.1 )-->for margins on top & bottom
    """
  #Define the axis biundary of the meshgrid & create a meshgrid
  x_max,x_min=x[:,0].min()-0.1 ,x[:,1].max()+0.1
  y_min,y_max=x[:,1].min()-0.1 ,x[:,1].max()+ 0.1

  xx,yy= np.meshgrid(np.linspace(x_min,x_max,100),
                     np.linspace(y_min,y_max,100))
  #Create the x values
  x_in=np.c_[xx.ravel(),yy.ravel()] #takes multiple 1D or 2D arrays and puts them side-by-side to form a bigger 2D array.

  #Make predidctions
  y_pred= model.predict(x_in)

  #Check for multiclass  prediction
  if (y_pred[0]>1):
    print("We're doing multiclass predictions")
    #Lets prepare the data for plotting
    y_pred=np.argmax(y_pred,axis=1).reshape(xx.shape)
  else:
    print("Doing binary classification")
    y_pred=np.round(y_pred).reshape(xx.shape)

  #Plot the decision biundary
  plt.contourf(xx,yy,y_pred, cmap=plt.cm.RdYlBu ,alpha=0.7)
  plt.scatter(x[:,0],x[:,1],s=40 ,c=y ,cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(),xx.max())
  plt.ylim(yy.min(),yy.max())
```

##How do you find the besst learning rate ??

To find the ideal learning rate (The learning rate at which the loss curve decreases the most )we use the following
1. A callback
2. Another model
3. A modified loss curve plot

```python
##
#Set sed
tf.random.set_seed(42
                   )

#01.Create the model
model_7=tf.keras.Sequential([
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(8,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

#02.Compile the modlel
model_7.compile(
    loss="binary_crossentropy",
    optimizer="Adam",
    metrics=["accuracy"]
)

#03.Use callback set up during training
lr_scheduler=tf.keras.callbacks.LearningRateScheduler(lambda epoch:1e-4*10**(epoch/20))

#04.Fit the model
history_7=model_7.fit(x_train,y_train,epochs=100,callbacks=[lr_scheduler])
```

```python
##Lets plot the loss curve and see
pd.DataFrame(history_7.history).plot(figsize=(12,6),xlabel="Epochs")
```

**Observation**: The loss decreases most/fastest around the (30-40)th epoch range.

```python
#Plotthin the learning rate vs the loss function
lrs=1e-4*10**(tf.range(100)/20)
lrs,len(lrs)
```

```python
plt.figure(figsize=(10,6))
plt.semilogx(lrs,history_7.history["loss"])
plt.xlabel("Learning rate")
plt.ylabel("Loss ")
plt.title("Learning rate VS Loss")
```

**Observation**: The ideal learning rate sits a slightly beyond 1e-2 to the lowest point touched by the curve

**Conclusions**: For our model ,the ideal learning rate is  around 1e-2

##Anatomy of confusion matrix
A confusion matrix is a table used in machine learning to summarize the results of a classification model, showing how many times it correctly or incorrectly predicted each class. It's a more detailed performance summary than a simple accuracy score, as it breaks down correct and incorrect predictions, and the types of errors being made. This helps you understand if a model consistently misclassifies a particular class.

1. True positive : Model predicts 1 when the truth is 1
2. True negative :Model predicts 0 when the truth is 0
3. False positive:Model predicts 1 when truth is 0
4. False negative: Model predicts 0 when truth is 1

##Note
When comparing models tweaking  should be gradually incremental so as to be able to figure out which tweaking caused what .

```python
##Lets try figuring out the ideal learning rate of the model

#set random seed
tf.random.set_seed(42)

#01.Create the model
model_11=tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(8,activation="relu"),
    tf.keras.layers.Dense(10,activation="softmax")
])

#02.Compile the model
model_11.compile(loss=tf.keras.losses.CategoricalCrossentropy,
                 optimizer=tf.keras.optimizers.Adam(),
                 metrics=["accuracy"])

#03.Create a callback
lr_scheduler=tf.keras.callbacks.LearningRateScheduler(lambda epochs:1e3*10**(epochs/20))

#04.Fit the model
lr_history=model_11.fit(train_data_norm,
             tf.one_hot(train_labels,depth=10),
             epochs=40,
             validation_data=(test_data,
                              tf.one_hot(test_labels,depth=10)),
             callbacks=[lr_scheduler])
```

```python
#Let's plot the Loss curve
import matplotlib.pyplot as plt

lrs=1e3*10**(tf.range(40)/20)
plt.semilogx(lrs,lr_history.history["loss"])
plt.xlabel("Learning rate")
plt.ylabel("Loss ")
plt.title("Finding Ideal L.R(Model_11)")
```
